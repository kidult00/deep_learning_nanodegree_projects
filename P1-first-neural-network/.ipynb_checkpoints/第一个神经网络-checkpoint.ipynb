# 你的第一个神经网络：预测共享单车的用户量

在这个项目中，你将搭建你的第一个神经网络，用来预测共享单车的租借数据。

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 加载数据和预处理

```python
data_path = 'Bike-Sharing-Dataset/hour.csv'
rides = pd.read_csv(data_path)
rides.head()
```

## 观察数据
这个数据集包含了 2011.1.1-2012.12.31 期间的每天，每一小时的乘客数据。乘客分为未注册和已注册两类，显示在 `cnt` 列。

下面的图表显示了数据集中前 10 天的乘客数（有些天数不包含完整的 24 个数据，所以并不是完整的 10 天）。你可以看到每小时的租借数。周末的租借数较少，工作日通勤时间出现高峰。上面的表格中，还包含了温度、湿度、风速等信息，它们也可能影响乘客数。

```python
rides[:24*10].plot(x='dteday', y='cnt');  # x= 表中的列名
```

### Dummy variables
表格中有一些类别变量如季节、天气、月份，如果想在模型中使用，需要转换为二进制的 dummy variables。用 Pandas 的 `get_dummies()` 就可以实现。

```python
dummy_fields = ['season', 'weathersit', 'mnth', 'hr', 'weekday']

for each in dummy_fields:
    dummies = pd.get_dummies(rides[each], prefix=each, drop_first=False) # 生成二进制列
    rides = pd.concat([rides, dummies], axis=1) # 沿 x 轴（axis=1，附加为列）附加在原来的数据表之后
    
fields_to_drop = ['instant', 'dteday', 'season', 'weathersit', 
                  'weekday', 'atemp', 'mnth', 'workingday', 'hr']  # 列出不显示的列名

data = rides.drop(fields_to_drop, axis=1)
data.head()
```

### Scaling target variables 归一化变量
为了得到更好的训练结果，需要「标准化」每一个连续变量，也就是说，把变量的范围统一成**均值为 0 ，标准差为 1 **。

我们把归一化的结果保存下来，方便在做预测的时候使用。

```python
quant_features = ['casual', 'registered', 'cnt', 'temp', 'hum', 'windspeed']

# 把归一化的结果存到一个字典里
scaled_features = {}

for each in quant_features:
    mean, std = data[each].mean(), data[each].std()
    scaled_features[each] = [mean, std]
    data.loc[:, each] = (data[each] - mean) / std  # 所有行，'名字'列：(数值-平均值)/标准差
    
scaled_features
```

归一化结果有问题。提示 py 3.6 .loc[] 用法

### 将数据集分为 training, testing, 和 validation 三部分

把最后 21 天数据用做测试，跟用模型生成的预测数据做对比。

```python
# 最后 21 天的数据划为测试集
test_data = data[-21*24:]

# 从训练数据中去掉测试集
data = data[:-21*24]

# 把数据分为输入（features）和输出（targets）
target_fields = ['cnt', 'casual', 'registered']
features, targets = data.drop(target_fields, axis=1), data[target_fields]
test_features, test_targets = test_data.drop(target_fields, axis=1), test_data[target_fields]
```

再把训练数据分为两部分，一部分用于训练，一部分用于验证。因为这是一个时间序列数据集，所以我们在历史数据上训练，在时间更新的验证集上面做验证。

```python
# 在训练集中划出最后 60 天作为验证集
train_features, train_targets = features[:-60*24], targets[:-60*24]
val_features, val_targets = features[-60*24:], targets[-60*24:]
```

```python
print('训练集输入:{} \n训练集输出:{}\n验证集输入:{} \n验证集输出:{}\n测试集输入:{}\n测试集输出:{}'
      .format(features.shape, targets.shape, val_features.shape, val_targets.shape, test_features.shape, test_targets.shape))
```

## 搭建神经网络

下面开始搭建神经网络，包括模型结构、backwards pass、forward pass，以及设置 hyperparameters: learning rate 学习率, hidden units 隐层节点数量, number of training 训练次数。

<img src="assets/neural_network.png" width=300px>

网络包含两层：隐层和输出层。隐层使用 sigmoid 函数作为激活函数。

>A function that takes the input signal and generates an output signal, but takes into account the threshold, is called an activation function

输出层只有一个节点，用于回归。输出节点跟输入节点一样，即激活函数是 $f(x)=x$. 

- ``正向传播 forward propagation``：计算网络中每个节点的输出，并且作为输入传递给下一层的神经元，即上一层节点的输出是下一层节点的输入。
- ``反向传播 backpropagation``：在神经网络中，使用权重从输入层传播 signals 到输出层。权重还被用于从输出层传播 error 到上一层，以便更新权重。

**思路**：

- 最终为了得到 △w，来更新 w，完成训练
- 为了得到 △w，需要输出层到隐层的△w: ``delta_weights_i_h``、隐层到输入层的△w: ``delta_weights_h_o``
- 为了得到两层的 △w，需要知道 隐层的输出 ``hidden_outputs``，输出层的输出 ``final_outputs``
- 为了得到两层的输出，需要知道输入的 feature


以下 class 实现：
1. Implement the sigmoid function to use as the activation function. Set `self.activation_function` in `__init__` to your sigmoid function.
2. Implement the forward pass in the `train` method.
3. Implement the backpropagation algorithm in the `train` method, including calculating the output error.
4. Implement the forward pass in the `run` method.
  

```python
class NeuralNetwork(object):
    def __init__(self, input_nodes, hidden_nodes, output_nodes, learning_rate):
        # 定义 输入层、隐层、输出层 节点数
        self.input_nodes = input_nodes
        self.hidden_nodes = hidden_nodes
        self.output_nodes = output_nodes
        # 定义学习率
        self.lr = learning_rate
        
        # 初始化输入层到隐层权重、隐层到输出层权重
        # normal(loc=0.0, scale=1.0, size=None)  - Draw random samples from a normal (Gaussian) distribution.
        self.weights_input_to_hidden = np.random.normal(0.0, self.input_nodes**-0.5,
                                                        (self.input_nodes, self.hidden_nodes))
        
        self.weights_hidden_to_output = np.random.normal(0.0, self.hidden_nodes**-0.5,
                                                        (self.hidden_nodes, self.output_nodes))

        # 激活函数：sigmoid 函数
        self.activation_function = lambda x : 1 / (1 + np.exp(-x))


    def train(self, features, targets):
        ''' Train the network on batch of features and targets.

            Arguments
            ---------

            features: 2D array, each row is one data record, each column is a feature
            targets: 1D array of target values

        '''
        # 输入数据个数
        n_records = features.shape[0]
        
        # 初始化权重步长 △w, 变量名与函数 backpropagation 的参数相同
        delta_weights_i_h = np.zeros(self.weights_input_to_hidden.shape)
        delta_weights_h_o = np.zeros(self.weights_hidden_to_output.shape)
        
        for X,y in zip(features, targets):
            # 调用 forword_pass_train 函数计算输出层和隐层的输出值
            final_outputs, hidden_outputs = self.forward_pass_train(X)
            # 调用 backpropagation 函数进行反向传播
            delta_weights_i_h, delta_weights_h_o = self.backpropagation(final_outputs, hidden_outputs, X, y, 
                                                                        delta_weights_i_h, delta_weights_h_o)
        # 调用 update_weights 函数更新权重
        self.update_weights(delta_weights_i_h, delta_weights_h_o, n_records)
            

    
    def forward_pass_train(self, X):
        ''' Implement forward pass here

            Arguments
            ---------
            X: features batch

        '''
        # 隐层输入：features * weights
        hidden_inputs = np.dot(X, self.weights_input_to_hidden)
        # 隐藏输出：激活函数（inputs）
        hidden_outputs = self.activation_function(hidden_inputs)
        
        # 输出层输出：features（隐层输出作为输出层的输入） * weights
        final_inputs = np.dot(hidden_outputs, self.weights_hidden_to_output)
        # 输出层不需要激活函数了
        final_outputs = final_inputs
        
        return final_outputs, hidden_outputs

    
    def backpropagation(self, final_outputs, hidden_outputs, X, y, delta_weights_i_h, delta_weights_h_o):
        ''' Implement backpropagation

            Arguments
            ---------
            final_outputs: output from forward pass
            y: target (i.e. label) batch
            delta_weights_i_h: change in weights from input to hidden layers
            delta_weights_h_o: change in weights from hidden to output layers

        '''

        # 计算输出层的误差，然后反向传播
        error = y - final_outputs
        
        # 反向传播误差到隐层
        # 权重需要转置：error 的 shape 是 (输出节点数) 行 1 列，隐层到输出层的 weight 是 （隐层节点数） 行 (输出节点数) 列
        hidden_error = np.dot(error, self.weights_hidden_to_output.T)
        
        # error_term = error * f'(x), 激活函数 sigmoid 的导数 x * (1 - x)
        hidden_error_term = hidden_error * hidden_outputs * (1 - hidden_outputs)
        # 输出层不需要激活函数
        output_error_term = error
        
        # 更新权重: error_term * input. [:, None] 用于转置， X * hidden_error_term 是否可以？
        delta_weights_i_h += hidden_error_term * X[:, None]
        delta_weights_h_o += output_error_term * hidden_outputs[:, None]
        
        return delta_weights_i_h, delta_weights_h_o
       

    def update_weights(self, delta_weights_i_h, delta_weights_h_o, n_records):
        ''' Update weights on gradient descent step

            Arguments
            ---------
            delta_weights_i_h: change in weights from input to hidden layers
            delta_weights_h_o: change in weights from hidden to output layers
            n_records: number of records

        '''
        # 梯度下降： 权重更新 = 学习率 * 权重步长 / 数据个数
        self.weights_hidden_to_output += self.lr * delta_weights_h_o / n_records
        self.weights_input_to_hidden += self.lr * delta_weights_i_h / n_records
        
        
    def run(self, features):
        ''' Run a forward pass through the network with input features

            Arguments
            ---------
            features: 1D array of feature values
        '''
        # 喂数据，得到 output，跟 forward_pass_train 函数的实现一样
        
        # 隐层输入：features * weights
        hidden_inputs = np.dot(features, self.weights_input_to_hidden)
        # 隐藏输出：激活函数（inputs）
        hidden_outputs = self.activation_function(hidden_inputs)
        
        # 输出层输出：features（隐层输出作为输出层的输入） * weights
        final_inputs = np.dot(hidden_outputs, self.weights_hidden_to_output)
        # 输出层不需要激活函数了
        final_outputs = final_inputs
        
        return final_outputs


# hyperparameters

iterations = 2000
learning_rate = 0.5
hidden_nodes = 10
output_nodes = 1
```

```python
# 均方误差（mean square error）
def MSE(y, Y):
    return np.mean((y-Y)**2)
```

## 单元测试

```python
import unittest

inputs = np.array([[0.5, -0.2, 0.1]])
targets = np.array([[0.4]])
test_w_i_h = np.array([[0.1, -0.2],
                       [0.4, 0.5],
                       [-0.3, 0.2]])
test_w_h_o = np.array([[0.3],
                       [-0.1]])

class TestMethods(unittest.TestCase):
    
    ##########
    # Unit tests for data loading
    ##########
    
    def test_data_path(self):
        # Test that file path to dataset has been unaltered
        self.assertTrue(data_path.lower() == 'bike-sharing-dataset/hour.csv')
        
    def test_data_loaded(self):
        # Test that data frame loaded
        self.assertTrue(isinstance(rides, pd.DataFrame))
    
    ##########
    # Unit tests for network functionality
    ##########

    def test_activation(self):
        network = NeuralNetwork(3, 2, 1, 0.5)
        # Test that the activation function is a sigmoid
        self.assertTrue(np.all(network.activation_function(0.5) == 1/(1+np.exp(-0.5))))

    def test_train(self):
        # Test that weights are updated correctly on training
        network = NeuralNetwork(3, 2, 1, 0.5)
        network.weights_input_to_hidden = test_w_i_h.copy()
        network.weights_hidden_to_output = test_w_h_o.copy()
        
        network.train(inputs, targets)
        self.assertTrue(np.allclose(network.weights_hidden_to_output, 
                                    np.array([[ 0.37275328], 
                                              [-0.03172939]])))
        self.assertTrue(np.allclose(network.weights_input_to_hidden,
                                    np.array([[ 0.10562014, -0.20185996], 
                                              [0.39775194, 0.50074398], 
                                              [-0.29887597, 0.19962801]])))

    def test_run(self):
        # Test correctness of run method
        network = NeuralNetwork(3, 2, 1, 0.5)
        network.weights_input_to_hidden = test_w_i_h.copy()
        network.weights_hidden_to_output = test_w_h_o.copy()

        self.assertTrue(np.allclose(network.run(inputs), 0.09998924))

suite = unittest.TestLoader().loadTestsFromModule(TestMethods())
unittest.TextTestRunner().run(suite)
```

## 训练网络

设置超参数，令训练集的 error 最小（但不至于过拟合）。如果网络训练得太久或者包含太多隐层节点，可能会过度拟合现有训练数据，没法应用到验证集，表现是训练 loss 下降，但是验证集的 loss 反而增加。

我们也可以使用 Stochastic Gradient Descent (SGD) 来训练神经网络。SGD 的思路是，每次训练都选取随机的样本，而不是整个数据集。用大量的小而快速的训练，代替完整的梯度下降，可以大大提高训练效率。

### 选取迭代次数 number of iterations
This is the number of batches of samples from the training data. 迭代次数越多，模型越准确。但是要注意迭代到一定程度，收益会骤减，以及需要大量的运算资源。理想的迭代次数，让训练在验证集的 loss 不再明显下降时停止。

### 选取学习率 learning rate
学习率主要调控权重更新的「幅度」。如果太大，可能无法找到合适的权重。**一般可以试着从 0.1 开始。** 如果模型没法拟合，先尝试降低学习率。学习率越低，权重调整的幅度就越小，当然训练时长也会相应增加。

### 选取隐层节点数 number of hidden nodes
在权重经过优化的模型中，隐层节点数越多，模型的预测结果会越准确。但是这也意味着更难优化权重（有太多的「学习维度」），或者会导致过拟合，在新的数据集上无法取得较好的效果。根据 loss 调整隐层节点数，**最适合的数量一般在输入节点数和输出节点数之间。**

```python
import sys

N_i = train_features.shape[1]
network = NeuralNetwork(N_i, hidden_nodes, output_nodes, learning_rate)

losses = {'train':[], 'validation':[]}

for ii in range(iterations):
    # SGD：从训练集随机选 128 个数据（的索引）放到 list 中作为 1 个 batch，注意 .index 的用法
    batch = np.random.choice(train_features.index, size=128)
    # 见下一个 cell .loc[] 用法
    X, y = train_features.loc[batch].values, train_targets.loc[batch]['cnt']
                             
    network.train(X, y)
    
    # Printing out the training progress
    # 训练集 loss. MSE(y,Y): return np.mean((y-Y)**2)
    train_loss = MSE(network.run(train_features).T, train_targets['cnt'].values)
    # 验证集 loss
    val_loss = MSE(network.run(val_features).T, val_targets['cnt'].values)
    
    sys.stdout.write("\rProgress: {:2.1f}".format(100 * ii/float(iterations)) \
                     + "% ... Training loss: " + str(train_loss)[:5] \
                     + " ... Validation loss: " + str(val_loss)[:5])
    sys.stdout.flush()
    
    losses['train'].append(train_loss)
    losses['validation'].append(val_loss)
```

``.loc[]`` is primarily label based, but may also be used with a boolean array.

Allowed inputs are:

- single label, e.g. 5 or 'a', (note that 5 is interpreted as a label of the index, and never as an integer position along the index).
- list or array of labels, e.g. ['a', 'b', 'c'].
- slice object with labels, e.g. 'a':'f' (note that contrary to usual python slices, both the start and the stop are included!).
- boolean array.
- callable function with one argument (the calling Series, DataFrame or Panel) and that returns valid output for indexing (one of the above)

```python
# np.random.choice?
# choice(a, size=None, replace=True, p=None)
# Generates a random sample from a given 1-D array
# Parameters
# -----------
# a : 1-D array-like or int
#     If an ndarray, a random sample is generated from its elements.
#     If an int, the random sample is generated as if a were np.arange(a)
```

```python
plt.plot(losses['train'], label='Training loss')
plt.plot(losses['validation'], label='Validation loss')
plt.legend()
_ = plt.ylim()
```
